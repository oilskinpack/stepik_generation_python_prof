```python
from datetime import datetime
from datetime import date
from datetime import time

#Создание
my_full_dt = datetime(1995,11,21,14,43) #С датой
my_part_dt = datetime(1995,11,21)       #Только с датой

print(my_full_dt)  #1995-11-21 14:43:00
print(my_part_dt)  #1995-11-21 00:00:00
```

```python
#Мето combine - объединяет Date и time
my_date = date(1992,10,6)
my_time = time(10,45,17)
my_datetime = datetime.combine(my_date,my_time)

print(my_datetime) #1992-10-06 10:45:17
```

```python
#Методы Date() и Time() получают экземпляры класса
print(my_datetime.date()) #1992-10-06
print(my_datetime.time()) #10:45:17
```

```python
#Метод now получает текущие дату время
from datetime import UTC

print(datetime.now())     #2025-03-11 18:49:18.193641 -мое время (+2 мск)
print(datetime.now(UTC))  #2025-03-11 13:49:18.193641+00:00 - время UTC (-3 мск)
```

```python
#Метод timestamp - количество секунд с момента нашей эпохи (1 января 1970)
#Метод fromtimestamp - наоборот
print(my_datetime.timestamp())  #718350317.0
print(datetime.fromtimestamp(323125454))  #1980-03-29 02:04:14
```

```python
from datetime import timedelta
data = {'Дима': ('03.11.2021 09:31:18', '03.11.2021 11:41:28'), 
        'Геор': ('01.11.2021 09:03:04', '01.11.2021 12:40:35'), 
        'Анна': ('02.11.2021 04:41:54', '02.11.2021 05:39:40'), 
        'Илина': ('02.11.2021 01:36:40', '02.11.2021 04:48:27'), 
        'Герман': ('04.11.2021 07:51:19', '04.11.2021 09:53:53'), 
        'Руслан': ('01.11.2021 11:26:06', '01.11.2021 12:56:24'), 
        'Лера': ('03.11.2021 11:09:41', '03.11.2021 14:37:41'), 
        'Егор': ('03.11.2021 05:29:38', '03.11.2021 06:01:59'), 
        'Максим': ('05.11.2021 13:05:03', '05.11.2021 14:27:41'), 
        'Саша': ('03.11.2021 04:14:26', '03.11.2021 05:10:58'), 
        'Марина': ('05.11.2021 15:21:06', '05.11.2021 18:33:46')}

winner = ''
min_time = timedelta(23,59,59)
for key,val in data.items():
    st,fin = val
    st = datetime.strptime(st,'%d.%m.%Y %H:%M:%S')
    fin = datetime.strptime(fin,'%d.%m.%Y %H:%M:%S')
    delta = fin - st
    if(delta < min_time):
        winner = key
        min_time = delta
print(winner)
```
