```python
from datetime import timedelta

delta = timedelta(days=7,hours=20,minutes=7)

print(delta)  #7 days, 20:07:00
```

```python
#При вводе значений больше макс доступного диапазона
delta1 = timedelta(days=50,weeks=2)

print(delta1) #64 days, 0:00:00
```

```python
#Отрицательное значение интервала
delta1 = timedelta(minutes=-40)

print(delta1) #-1 day, 23:20:00
```

```python
#Timedelta не содержит Hours и Minutes
def hours_minutes(td):
    return td.seconds // 3600, (td.seconds // 60) % 60

delta = timedelta(days=7, seconds=125, minutes=10, hours=8, weeks=2)

hours, minutes = hours_minutes(delta)

print(delta) #21 days, 8:12:05
print(hours) #8
print(minutes) #12
```

```python
#Сравнение временных интервалов
delta1 = timedelta(weeks=3,hours=7)
delta2 = timedelta(weeks=3,minutes=13)
delta3 = timedelta(weeks=3)

print(delta1 > delta2) #True
print(delta1 == 3)
print(delta3 == 3)
```

```python
#Операции над временными интервалами
#Сложение вычитание
delta1 = timedelta(days=5) + timedelta(seconds=3600)
delta2 = timedelta(days=5) - timedelta(seconds=3600)

print(delta1) #5 days, 1:00:00
print(delta2) #4 days, 23:00:00

#Умножение и деление на число
delta3 = timedelta(days=1) / 24
delta4 = timedelta(days=1) * 7

print(delta3) #1:00:00
print(delta4) #7 days, 0:00:00

#Деление интервала на интервал
delta5 = timedelta(weeks=1) / timedelta(hours=5) 

print(delta5)  #33.6
```

```python
def num_of_sundays(year):
    counter = 0
    first_day = datetime(year=year,month=1,day=1)
    last_day = first_day.replace(month=12,day=31)

    for day in range(first_day.toordinal(),last_day.toordinal()+1):
        this_date = datetime.fromordinal(day)
        if(this_date.weekday() == 6):
            counter += 1

    return counter



year = 2000
counter = 0
first_day = datetime(year=year,month=1,day=1)
last_day = first_day.replace(month=12,day=31)

for day in range(first_day.toordinal(),last_day.toordinal()+1):
    this_date = datetime.fromordinal(day)
    if(this_date.weekday() == 6):
        counter += 1

print(counter)
```

```python
def fill_up_missing_dates(dates):
    # dates = ['01.11.2021', '07.11.2021', '04.11.2021', '03.11.2021']
    dates_dt = list(map(lambda x: datetime.strptime(x,'%d.%m.%Y'),dates))
    min_date = min(dates_dt)
    max_date = max(dates_dt)

    res = []
    for d in range(min_date.toordinal(),max_date.toordinal()+1):
        res.append(datetime.fromordinal(d).strftime('%d.%m.%Y'))
    return res
```
