```python
#Год и квартал
dates = [date(2010, 9, 28), date(2017, 1, 13), date(2009, 12, 25),
          date(2010, 2, 27), date(2021, 10, 11), date(2020, 3, 13),
            date(2000, 7, 7), date(1999, 4, 14), date(1789, 11, 19),
              date(2013, 8, 21), date(1666, 6, 6), date(1968, 5, 26)]
for d in dates:
    if d.month <= 3:
        print(f'{d.year}-Q1')
    elif d.month <= 6:
        print(f'{d.year}-Q2')
    elif d.month <= 9:
        print(f'{d.year}-Q3')
    else:
        print(f'{d.year}-Q4')
```

```python
#Функция возвращает кортеж макс и мин дат
def get_min_max(dates):
    if(len(dates) == 0):
        return tuple()
    min_date = min(dates)
    max_date = max(dates)
    return tuple([min_date,max_date])

get_min_max([date(1994,4,5),date(1996,4,5),date(1993,4,5)])
```

```python
#Функция возвращает список всех дат от мин до макс включительно
def get_date_range(start,end):
    res = []
    if(end < start):
        return res
    for d in range(start.toordinal(),end.toordinal()+1):
        res.append(date.fromordinal(d))
    return res
my_st = date(1994,4,5)
my_fn = date(1994,4,7)
get_date_range(start=my_st,end=my_fn)
```

```python
#Функция выводящая количество суббот в диапазоне
def saturdays_between_two_dates(start,end):
    res = []
    min_date = min([start,end])
    max_date = max([end,start])
    for d in range(min_date.toordinal(),max_date.toordinal()+1):
        this_date = date.fromordinal(d)
        day_num = this_date.weekday()
        if(day_num == 5):
            res.append(this_date)
    return len(res)

my_st = date(2025,3,1)
my_fn = date(2025,3,9)
saturdays_between_two_dates(start=my_st,end=my_fn)
```

```python
from datetime import date
def print_good_dates(dates):
    interesting_dates = []
    if (len(dates)==0):
        return None

    for d in dates:
        year = d.year
        month = d.month
        day = d.day
        birthday_day =  month + day
        if(year == 1992 and birthday_day == 29):
            interesting_dates.append(d)
    srt_dates = sorted(interesting_dates)
    for d in srt_dates:
        print(d.strftime('%B %d, %Y'))

dates = []
print_good_dates(dates)
```

```python
def is_correct(day,month,year):
    try:
        date(year=year,month=month,day=day)
        return True
    except:
        return False
print(is_correct(31, 12, 2021))
```
