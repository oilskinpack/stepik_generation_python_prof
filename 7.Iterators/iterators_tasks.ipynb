```python
class Repeater():

    def __init__(self,obj):
        self.obj = obj

    def __iter__(self):
        return self
    
    def __next__(self):
        return self.obj

bee = Repeater('bee')

print(next(bee))
```

```python
class BoundedRepeater:

    def __init__(self,obj,times):
        self.obj = obj
        self.times = times
        self.iteration = 1
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.iteration <= self.times:
            self.iteration += 1
            return self.obj
        
        raise StopIteration
    
geek = BoundedRepeater('geek', 3)

print(next(geek))
print(next(geek))
print(next(geek))

try:
    print(next(geek))
except StopIteration:
    print('Error')
```

```python
class Square:

    def __init__(self,n):
        self.n = n
        self.value = 1
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.n >= self.value:
            next_value = self.value ** 2
            self.value +=1
            return next_value
        raise StopIteration
    
squares = Square(1)

print(list(squares))
```

```python
class Fibonacci:

    def __init__(self):
        self.first = 0
        self.second = 1
        self.iter = 0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.iter == 0:
            self.iter += 1
            return 1
        
        res = self.first + self.second
        prev = self.second
        self.second = res
        self.first = prev
        return res
    
fibonacci = Fibonacci()

print(next(fibonacci))
print(next(fibonacci))
print(next(fibonacci))
print(next(fibonacci))
```

```python
class DictItemsIterator:
    def __init__(self, data):
        self.data = list(data.items())  #      -
        self.index = 0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.index < len(self.data):
            result = self.data[self.index]
            self.index += 1
            return result
        raise StopIteration
        
pairs = DictItemsIterator({1: 'A', 2: 'B', 3: 'C'})

print(*pairs)
```

```python
class CardDeck:
    def __init__(self):
        self.masti = ("", "", "", "")
        self.numbers = ("2", "3", "4", "5", "6", "7", "8", "9", "10", "", "", "", "")
        self.masti_ind = 0
        self.numbers_ind = 0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        try:
            if self.numbers_ind > 12:
                self.masti_ind += 1
                self.numbers_ind = 0
            res = f'{self.numbers[self.numbers_ind]} {self.masti[self.masti_ind]}'
            self.numbers_ind += 1
            return res
        except:
            raise StopIteration
        
cards = list(CardDeck())

print(cards[9])
print(cards[23])
print(cards[37])
print(cards[51])
```

```python
class Cycle:
    def __init__(self,iterable):
        self.data = iterable
        self.iter = 0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.iter < len(self.data):
            res = self.data[self.iter]
            self.iter += 1
            return res
        self.iter = 0
        res = self.data[self.iter]
        self.iter += 1
        return res
    
cycle = Cycle(range(100_000_000))

print(next(cycle))
print(next(cycle))
```
