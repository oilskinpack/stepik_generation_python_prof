```python
range_obj = range(3)

print(next(range_obj)) #'range' object is not an iterator

#Range - это не итератор!!!
```

```python
range_obj = range(3)

print(list(range_obj)) #[0, 1, 2]
print(list(range_obj))  #[0, 1, 2]

#Элементы в Range не пропадают после использования
```

```python
range_obj = range(5, 50, 5)

print(*range_obj)
print(len(range_obj))
print(range_obj[3])
print(range_obj[-1])
print(range_obj[4:8])

#Поддерживает индексацию и срезы
```

```python
range_obj = range(5, 50, 5)

print(15 in range_obj)
print(15 in range_obj)
print(15 in range_obj)

#Можем определять вхождения
```

```python
range1 = range(10)
range2 = range(0, 10, 1)
range3 = range(5, 50, 10)

print(range1 == range2)
print(range1 == range3)

#Можно сравнивать ренжы
```

```python
from collections import Counter

def unique(iterable):
    return iter(Counter(iterable))

numbers = [1, 2, 2, 3, 4, 5, 5, 5]

print(*unique(numbers))
```

```python
def stop_on(iterable,obj):
    for i in iterable:
        if(i == obj):
            return
        yield i

numbers = [1, 2, 3, 4, 5]

print(*stop_on(numbers, 4))
```

```python
def with_previous(iterable):
    prev = None
    for i in iterable:
        yield (i,prev)
        prev = i

numbers = []

print(*with_previous(numbers))
```

```python
def pairwise(iterable):
    iterator = iter(iterable)
    try:
        prev = next(iterator)
    except StopIteration:
        return
    
    for item in iterator:
        yield (prev, item)
        prev = item
    yield (prev, None)

data = map(str.upper, 'jhfjgshgkjfdjsgriyteryowpqerkelfsldfmnmnbmvcnmlgqweootiyoeytkldjhmvxcmkasd')

print(*pairwise(data))
```

```python
def around(iterable):
    first = None
    iterator = iter(iterable)
    try:
        prev = next(iterator)
    except StopIteration:
        return
    
    for item in iterator:
        yield (first,prev, item)
        first = prev
        prev = item
    yield (first,prev, None)

numbers = [1, 2, 3, 4, 5]

print(*around(numbers))
```
