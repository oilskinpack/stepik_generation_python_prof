```python
def sum_digits(number):
    if number < 10:
        return number
    x = number % 10
    return x + sum_digits(number // 10)

sum_digits(341)
```

```python
def count_frogs(year):
    if year == 1:
        return 77 # 
    return 3 * (count_frogs(year - 1) - 30) # 

count_frogs(45)
```

```python
def range_sum(numbers, start, end):
    if start == end:
        return numbers[start]
    return numbers[start] + range_sum(numbers, start + 1, end)


print(range_sum([21, 2342, 34, 47, 5, 65, 47, 855, 9, 123, 42, 5, 6, 765, 34, 3, 32], 0, 1))
```

```python
def get_pow(a,n):
    if n == n:
        return a ** n
    return a * get_pow(a,n - 1)

get_pow(5, 3)
```

```python
def get_fast_pow(a,n):
    if n == 1:
        if(a // 2 == 0):
            return (a ** 2) ** (n /2)
        else:
            return a * (a ** (n - 1))
    return a * get_fast_pow(a,n-1)

print(get_fast_pow(2, 1000))
```

```python
def is_power(number):
    if number == 2:
        return True
    elif number < 2 and number == 1:
        return True
    elif number < 2:
        return False
    return is_power(number / 2)

is_power(17)
```

```python
def fib(n):
    dick = {1: 1, 2: 1,3:1}
    def fib_rec(n):
        result = dick.get(n)
        if result is None:
            result = fib_rec(n - 3) + fib_rec(n - 2) + fib_rec(n - 1)
            dick[n] = result
        return result
    return fib_rec(n)
fib(5)
```

```python
def to_binary(number):
    if number == 0:
        return "0"
    if number == 1:
        return "1"
    part_one = to_binary(number // 2)
    part_two = str(number % 2)
    return part_one + part_two

to_binary(7)
```

```python
def recursive_sum(nested_lists):
    total = 0
    def rec(nested_lists,total):
        if type(nested_lists) == int:
            total += nested_lists    #  
        if type(nested_lists) == list:
            for i in nested_lists:
                rec(i,total)              #  
    rec(nested_lists,total)
    return total

my_list = [1, [4, 4], 2, [1, [2, 10]]]

print(recursive_sum(my_list))
```

```python
def recursive_sum(nested_lists):
    total = 0
    if isinstance(nested_lists, list):
        for element in nested_lists:
            total += recursive_sum(element)
    elif isinstance(nested_lists, int):
        total += nested_lists
    return total
```

```python
def linear(nested_lists):
    total = []
    if isinstance(nested_lists, list):
        for element in nested_lists:
            linear(element)
    else:
        total.append(nested_lists)
    return total

my_list = [3, [4], [5, [6, [7, 8]]]]

print(linear(my_list))
```

```python
def linear(nested_lists):
    total = []
    for element in nested_lists:
        if isinstance(element, list):
            total.extend(linear(element))
        else:
            total.append(element)
    return total
```

```python
def get_value(nested_dicts,key):
    if key in nested_dicts:
        return nested_dicts[key]
    
    for v in nested_dicts.values():
        if(type(v) == dict):
            value = get_value(v,key)
            if value is not None:
                return value
```

```python
def get_all_values(nested_dicts,key):
    res_set = set()
    if key in nested_dicts:
        res_set.add(nested_dicts[key])
    
    for v in nested_dicts.values():
        if(type(v) == dict):
            get_all_values(v,key)
    return res_set

my_dict = {'users': {'Arthur': {'grades': [4, 4, 3], 'top_grade': 4}, 'Timur': {'grades': [5, 5, 5], 'top_grade': 5}}}
result = get_all_values(my_dict, 'top_grade')
result
```

```python
def get_all_values(nested_dicts, key):
    res_set = set()
    
    if key in nested_dicts:
        res_set.add(nested_dicts[key])
    
    for v in nested_dicts.values():
        if isinstance(v, dict):
            res_set.update(get_all_values(v, key))
    
    return res_set

```

```python
def dict_travel(nested_dicts, path=None):
    if path is None:
        path = []
    
    for key in sorted(nested_dicts.keys()):
        value = nested_dicts[key]
        current_path = path + [key]
        
        if isinstance(value, dict):
            dict_travel(value, current_path)
        else:
            print(".".join(current_path) + ": " + str(value))

data = {'a': 1, 'b': {'c': 30, 'a': 10, 'b': 20}}

dict_travel(data)
```
