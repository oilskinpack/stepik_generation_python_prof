```python
import pandas as pd

df = pd.read_csv('meetings.csv',sep=',')
df['meeting'] = df['meeting_date'] + ' ' +  df['meeting_time']
df['meeting_datetime'] = pd.to_datetime(df['meeting'])
df_res = df.sort_values('meeting_datetime',ascending=True)[['surname','name']]
[print(f"{n['surname']} {n['name']}") for _,n in df_res.iterrows()]
```
