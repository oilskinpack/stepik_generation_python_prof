```python
data = [('Тимур', 'Артур'), ('Тимур', 'Дима'), ('Дима', 'Артур')]


def wins(pairs):
    history = defaultdict(set)
    for winner,loser in pairs:
        history[winner].add(loser)
    return history
wins(data)
```

```python
data = {'a': [1, 2], 'b': [3, 1], 'c': [2]}


def flip_dict(dict_of_lists):
    flip = defaultdict(list)
    for key,value in dict_of_lists.items():
        for v in value:
            flip[v].append(key)
    return flip
flip_dict(data)
```
