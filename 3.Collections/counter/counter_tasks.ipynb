```python
test = ',,,,, ,,,'

counter = dict(Counter(test.split(',')))
res = sorted(counter.items(),key= lambda x: x[0])
max_len = len(max(counter.keys(),key=lambda x:len(x)))
for x in res:
    name = x[0]
    price = 0
    spaces_counter = max_len - len(name)
    spaces = ' ' * spaces_counter
    for ch in x[0]:
        if(ch != ' '):
            price += ord(ch)
    full_price = price * x[1]
    print(f'{name}{spaces}: {price} UC x {x[1]} = {full_price} UC')
```

```python
from collections import Counter

def count_letters():
    with open('pythonzen.txt', encoding='utf-8') as f:
        text = f.read().lower()

    #   
    letters_only = [char for char in text if char.isalpha()]
    
    #  
    counter = Counter(letters_only)

    #    
    for letter in sorted(counter):
        print(f"{letter}: {counter[letter]}")

count_letters()
```

```python
def scrabble(symbols,word):
    symbols = symbols.lower()
    word = word.lower()

    sym_counter = Counter(symbols)
    w_counter = Counter(word)

    sym_counter.subtract(w_counter)
    lst =  list(sym_counter.values())
    for x in lst:
        if(int(x)) < 0:
            return False
    return True

print(scrabble('begk', 'beegeek'))
```

```python
import csv
import json

#  
with open('prices.json', encoding='utf-8') as f:
    prices = json.load(f)

#    
quarters = ['quarter1.csv', 'quarter2.csv', 'quarter3.csv', 'quarter4.csv']

#  
total = 0

#   CSV-
for filename in quarters:
    with open(filename, encoding='utf-8') as file:
        reader = csv.reader(file)
        next(reader)  #  

        for row in reader:
            product = row[0]
            amounts = map(int, row[1:])  #   
            total_kg = sum(amounts)
            price = prices[product]
            total += total_kg * price

print(total)
```

```python
from collections import Counter

#   
stock = list(map(int, input().split()))
books = Counter(stock)

#  
n = int(input())

total = 0

for _ in range(n):
    grade, offer = map(int, input().split())
    if books[grade] > 0:
        total += offer
        books[grade] -= 1  #    

print(total)
```
