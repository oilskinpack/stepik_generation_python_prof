```python
from datetime import datetime
d = {}
res = []
with open('name_log.csv',encoding='utf-8') as file:
    data = csv.reader(file,delimiter=',')
    next(data)
    logs = [row for row in data]
    sorted_logs = sorted(logs,key=lambda x:  datetime.strptime(x[2], '%d/%m/%Y %H:%M'),reverse=True)
    for log in sorted_logs:
        if(not(log[1] in d.keys())):
            d[log[1]] = [log[0],log[2]]
    sorted_d = sorted(d.items(),key=lambda item: item[0])

    
    for key,value in sorted_d:
        res.append([value[0],key,value[1]])
    
with open('new_name_log.csv','w',encoding='utf-8',newline='') as file:
    cols = ['username','email','dtime']
    writer = csv.writer(file)
    writer.writerow(cols)
    for row in res:
        writer.writerow(row)
```

```python
def condense_csv(filename,id_name):
    save_name = 'condensed.csv'
    columns = [id_name] 
    d = {id_name:[]}
    with open(filename,encoding='utf-8') as file:
        data = list(csv.reader(file,delimiter=','))
        #Собираем колонки
        first_id = data[0][0]
        for row in data:
            if(row[0] == first_id):
                d[row[1]] = []

        #Добавляем значения
        for row in data:
            if(not(row[0] in d[id_name])):
                d[id_name].append(row[0])
            d[row[1]].append(row[2])

    #Запись
    with open(save_name,'w',encoding='utf-8',newline='') as file:
        cols = d.keys()
        writer = csv.writer(file)
        writer.writerow(cols)
        m = [list(r) for r in zip(*d.values())]
        for row in m:
            writer.writerow(row)
```
