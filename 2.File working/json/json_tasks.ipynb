```python
with open(r'data_2.json',encoding='utf-8') as file:
    d = []
    data = json.load(file)
    for it in data:
        tp = type(it)
        if(tp is str):
            d.append(it + '!')
        elif(tp is int or tp is float):
            d.append(it + 1)
        elif(tp is bool):
            d.append(not(it))
        elif(tp is list):
            d.append(it * 2)
        elif(tp is dict):
            it['newkey'] = None
            d.append(it)
        

with open(r'updated_data.json','w',encoding='utf-8',newline='') as file:
    data = json.dump(d,file,indent=3)
    print(d)
```

```python
with open(r'people.json',encoding='utf-8') as file:
    all_keys = []
    new_data = []
    
    data = json.load(file)
    for man in data:
        [all_keys.append(k) for k in man.keys()]
    for man in data:
        for k in all_keys:
            if(not(k in man.keys())):
                man[k] = None
        new_data.append(man)
        # print(man)
    

with open(r'updated_people.json','w',encoding='utf-8',newline='') as file:
    json.dump(new_data,file,indent=3)
```

```python
import json
from collections import defaultdict

#    input_file
with open(r'countries2.json', 'r', encoding='utf-8') as f:
    countries = json.load(f)
    religion_map = defaultdict(list)

#    
for item in countries:
    religion = item['religion']
    country = item['country']
    religion_map[religion].append(country)
#  defaultdict     
with open('religion.json', 'w', encoding='utf-8') as f:
    json.dump(dict(religion_map), f, ensure_ascii=False, indent=3)
```

```python
my_students = []
with open(r'students.json',encoding='utf-8') as file:
    rows = json.load(file)
    for row in rows:
        age = int(row['age'])
        progress = int(row['progress'])
        if(age >= 18 and progress >= 75):
            my_students.append({'name':row['name'],'phone':row['phone']})
res = sorted(my_students,key=lambda x: x['name'])
with open(r'data.csv','w',newline='') as file:
    cols = ['name','phone']
    writer = csv.DictWriter(file, fieldnames=cols,delimiter=',')
    #      
    writer.writeheader()
    for row in res:
        writer.writerow(row)
# print(res)
```

```python
# with open(r'pools.json',encoding='utf-8') as file:
#     data = json.load(file)
#     for pool in data:
#         print(pool)
import pandas as pd
df = pd.read_json('pools.json')
df['Monday_Working'] = df['WorkingHoursSummer'].apply(lambda x:x[''])
df['Length'] = df['DimensionsSummer'].apply(lambda x: x['Length'])
df['Width'] = df['DimensionsSummer'].apply(lambda x: x['Width'])
df['Monday_Start'] = df['Monday_Working'].str.split('-').str[0].str.split(':').str[0]
df['Monday_End'] = df['Monday_Working'].str.split('-').str[1].str.split(':').str[0]

my_pools = df[(df['Monday_Start'].astype(float) <= 10) 
         & (df['Monday_End'].astype(float) >= 12)]
need_pool = my_pools.sort_values(by=['Length','Width'],ascending=[False,False]).iloc[0]
length = int(need_pool['Length'])
width = int(need_pool['Width'])
address = need_pool['Address']
print(f'{length}x{width}\n{address}')
```

```python
import json

#  JSON-
with open('food_services.json', encoding='utf-8') as f:
    data = json.load(f)  # ,    

#    
type_max_place = {}

for item in data:
    type_obj = item.get('TypeObject')
    name = item.get('Name')
    seats = item.get('SeatsCount')

    #       
    if type_obj and name and isinstance(seats, (int, float)):
        #             SeatCount
        if type_obj not in type_max_place or seats > type_max_place[type_obj]['SeatsCount']:
            type_max_place[type_obj] = {
                'Name': name,
                'SeatsCount': seats
            }

#    [(TypeObject, dict), ...]     
sorted_places = sorted(type_max_place.items(), key=lambda x: x[0])

#  
for type_obj, info in sorted_places:
    print(f"{type_obj}: {info['Name']}, {info['SeatsCount']}")

```
