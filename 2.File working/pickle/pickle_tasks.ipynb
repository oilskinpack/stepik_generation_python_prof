```python
def filter_dump(filename,objects,typename):
    with open(filename,'wb') as file:
        sorted_lst = [obj for obj in objects if type(obj) == typename]
        pickle.dump(sorted_lst,file)
        print('Успех')

filter_dump('numbers.pkl', [1, '2', 3, 4, '5'], int)

#Загрузка из pkl
with open('numbers.pkl','rb') as file:
    obj = pickle.load(file)
    print(obj)       #{'Python': 1991, 'Java': 1995, 'C#': 2002}  
```

```python
import pickle
import sys

# Читаем входные данные
file_name = input().strip()
control_value = int(input().strip())

# Загружаем объект из pickle-файла
with open(file_name, 'rb') as f:
    obj = pickle.load(f)

# Вычисление контрольной суммы
checksum = 0

if isinstance(obj, list):
    int_items = [i for i in obj if isinstance(i, int)]
    if int_items:
        checksum = min(int_items) * max(int_items)
    else:
        checksum = 0

elif isinstance(obj, dict):
    int_keys = [k for k in obj if isinstance(k, int)]
    checksum = sum(int_keys)
else:
    # Нестандартный формат — просто 0
    checksum = 0

# Сравнение и вывод результата
if checksum == control_value:
    print("Контрольные суммы совпадают")
else:
    print("Контрольные суммы не совпадают")
```
