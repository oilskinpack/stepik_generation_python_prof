```python
def numbers_sum(elems):
    """       (int, float),\n  . 0 -     ."""
    nums_only = list(filter(lambda x: isinstance(x,(int,float)),elems))
    if len(nums_only) > 0:
        return sum(nums_only)
    return 0

print(numbers_sum([1, '2', 3, 4, 'five']))
print(numbers_sum.__doc__)
```

```python
orig = print

def my_print(*args,sep=' ',end=''):
    new_args = [n.upper() if isinstance(n,str) else n for n in args]
    print(*new_args,sep=sep.upper(),end=end.upper())

print('bee', 'geek', sep=' and ', end=' wow')
```

```python
def countries(country, capital):
    countries.__dict__[country] = capital
    
countries.__dict__['Germany'] = 'Berlin'
countries.Norway = 'Oslo'
countries('Finland', 'Helsinki')

print(countries.__dict__)
```

```python
def polynom(x):
    res = x ** 2 + 1
    if not hasattr(polynom, 'values'):
        polynom.values = set()
        polynom.values.add(res)
    else:
        old_v = polynom.values
        old_v.add(res)
    return res

print(polynom(5))
print(polynom.values)
```
