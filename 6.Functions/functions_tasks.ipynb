```python
def is_greater(lists,number):
    return any(list(map(lambda x: sum(x) > number,lists)))

data = [[-3, 4, 0, 1], [1, 1, -4], [0, 0], [9, 3]]

print(is_greater(data, 10))
```

```python
def zip_longest(*args,fill=None):
    all_lists = [*args]
    max_len = len(max(all_lists,key=lambda x:len(x)))

    filled_lists = []
    for l in all_lists:
        dev = max_len - len(l)
        if(dev == 0):
            filled_lists.append(l)
        else:
            l.extend([fill]*dev)
            filled_lists.append(l)
    return list(zip(*args))

print(zip_longest([1, 2, 3, 4, 5], ['a', 'b', 'c'], fill='_'))
```

```python
def custom_sort(s):
    lower = sorted([c for c in s if c.islower()])
    upper = sorted([c for c in s if c.isupper()])
    odd_digits = sorted([c for c in s if c.isdigit() and int(c) % 2 == 1])
    even_digits = sorted([c for c in s if c.isdigit() and int(c) % 2 == 0])
    
    return ''.join(lower + upper + odd_digits + even_digits)

# Пример:
input_str = input()
print(custom_sort(input_str))
```

```python
def hash_as_key(objects):
    dick = {}
    for i in objects:
        if hash(i) in dick.keys():
            old_v = dick[hash(i)]
            if(isinstance(old_v,list)):
                old_v.append(i)
            else:
                dick[hash(i)] = [old_v,i]
        else:
            dick[hash(i)] = i
    return dick

data = [-1, -2, -3, -4, -5]

print(hash_as_key(data))
```
