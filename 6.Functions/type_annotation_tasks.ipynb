```python
NumOnly = int | float

def get_digits(number: NumOnly) -> list[int]:
    def try_parse(text):
        try:
            return int(text)
        except:
            return None
    nums_only = list(filter(lambda x: isinstance(x,int),[try_parse(num) for num in str(number)]))
    return nums_only

print(get_digits(16733))

annotations = get_digits.__annotations__
print(annotations)
```

```python


def top_grade(grades:dict[str,str | list[int]]) -> dict[str,int | str]:
    res = {}
    res['name'] = grades['name']
    res['top_grade'] = max(grades['grades'])
    return res

print(top_grade.__annotations__)
#dict[str, str | list[int]] dict[str, str | int]

```

```python
def cyclic_shift(numbers: list[int | float], step: int) -> None:
    n = len(numbers)
    if n == 0:
        return
    step = step % n  #  
    numbers[:] = numbers[-step:] + numbers[:-step]
```

```python
def matrix_to_dict(matrix: list[list[int | float]]) -> dict[int, list[int | float]]:
    return {i + 1: row for i, row in enumerate(matrix)}

matrix = [[5, 6, 7], [8, 3, 2], [4, 9, 8]]

print(matrix_to_dict(matrix))
```
