```python
def power(degree):
    def inner(x):
        return x ** degree
    return inner
square = power(2)
print(square(5))
```

```python
def generator_square_polynom(a,b,c):
    def inner(x):
        return a * (x**2) + b * x + c
    return inner

polyn = generator_square_polynom(1,1,1)
print(polyn(3))
```

```python
def sourcetemplate(url):
    def inner(**dick):
        # dick = {'Имя':"Никита",'Фамилия':'Хабаров'}
        # url = 'https://beegeek.ru'
        if len(dick) == 0:
            return url
        else:
            vals_lst = sorted([f'{k}={v}' for k,v in dick.items()],key=lambda x: x)
            vals_str = '&'.join(vals_lst)
            return f'{url}?{vals_str}'
    return inner

url = 'https://all_for_comfort_life.com'
load = sourcetemplate(url)
print(load(smartphone='iPhone', notebook='huawei', sale=True))
#https://all_for_comfort_life.com?notebook=huawei&sale=True&smartphone=iPhone

```

```python
from datetime import date

def date_formatter(country_code):
    def inner(my_date):
        # my_date = date(year,month,day)
        if country_code == 'ru':
            return my_date.strftime('%d.%m.%Y')
        elif country_code == 'us':
            return my_date.strftime('%m-%d-%Y')
        elif country_code == 'ca':
            return my_date.strftime('%Y-%m-%d')
        elif country_code == 'br':
            return my_date.strftime('%d/%m/%Y')
        elif country_code == 'fr':
            return my_date.strftime('%d.%m.%Y')
        else:
            return my_date.strftime('%d-%m-%Y')
    return inner
date_ru = date_formatter('ru')
today = date(2022, 1, 25)
print(date_ru(today))
```
